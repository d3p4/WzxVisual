```python
#实践1：
#导入绘图相关模块
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  #设置中文字体

#生成数据
month = ['1月','2月','3月','4月','5月','6月','7月','8月','9月','10月','11月','12月']
sales = [200,197,233,216,345,319,197,357,376,416,355,408]

#设置图形大小
plt.figure(figsize=(11,7))

#添加点，设置点的样式，颜色，大小，线宽
plt.plot(month,sales,linestyle='-.',marker='*',color='green',markersize=10,linewidth=3.0)

#设置坐标刻度值的大小以及刻度值的字体大小
plt.tick_params(labelsize=16)

#添加坐标标签
plt.xlabel('月份',size=16)
plt.ylabel('订单量',size=16)

#添加标题
plt.title('2019年12个月某企业商品的订单量分析',loc='center', size=20)

#输出图形
plt.show()
```

```python
#实践2：
#声明Notebook类型，必须在引入Pyecharts.charts等模块前声明
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB

from pyecharts.charts import Bar
from pyecharts import options as opts

bar = (
    Bar()
    .add_xaxis(['1月','2月','3月','4月','5月','6月','7月','8月','9月','10月','11月','12月'])
    .add_yaxis("订单量",[200,197,233,216,345,319,197,357,376,416,355,408])
    .set_global_opts(title_opts=opts.TitleOpts(title="2019年12个月商品的订单量",title_textstyle_opts=opts.TextStyleOpts(font_size=20)),
                     xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(font_size = 16)),
                     yaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(font_size = 16)),
                     toolbox_opts=opts.ToolboxOpts(),
                     legend_opts=opts.LegendOpts(is_show=True,item_width=40,item_height=20,textstyle_opts=opts.TextStyleOpts(font_size=16)))
    .set_series_opts(label_opts=opts.LabelOpts(font_size = 16))
)

#第一次渲染时候调用load_javasrcript文件
bar.load_javascript()
bar.render_notebook()
```

```python
#实践3：
#导入相关库
import plotly.offline as py
import plotly.graph_objects as go

month = ['第一季度','第二季度','第三季度','第四季度']
sales = [630,880,930,1179]

fig = go.Figure(data=[go.Pie(labels=month, values=sales, textinfo='label+percent',insidetextorientation='radial')])

fig.update_layout(
    title = dict(
       text = "2019年季度销售业绩分析",   #标题文本
       x = 0.47,                        #横向的相对位置，范围0-1
       y = 0.95                         #纵向的相对位置，范围0-1
    ),
    font=dict(
        family = "Microsoft Yahei UI",  #文字的字体
        size = 20,                      #文字的大小
        color = 'black'                 #文字的颜色      
    ),
    xaxis_title="月份",                      #X轴标签
    xaxis_title_font_family='MS Sans Serif', #设置x轴标签的字体
    xaxis_title_font_size=16,                #设置x轴标签的大小
    xaxis_title_font_color = 'red',          #设置x轴标签的颜色  
    yaxis_title="业绩数据",                   #Y轴标签
    yaxis_title_font_family='MS Sans Serif', #设置x轴标签的字体
    yaxis_title_font_size=16,                #设置x轴标签的大小
    yaxis_title_font_color = 'red'           #设置x轴标签的颜色     
)

py.plot(fig, filename='pie.html')
```
